```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# step 2 = Create the data frame
```

```python
dataframe = pd.read_csv("zomato_data.csv")
```

```python
print(dataframe)
```

```python
dataframe
```

```python
# data cleaning
```

```python
#convert the data type of column rate
```

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
    
```

```python
dataframe.info()
```

# Type of Resturant

```python
dataframe.head()
```

```python
# only for counting type of restaurant used here countplot
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of Restaurant")
```

```python
#conclusion= majority of the restaurant falls in dining category
```

#### how many votes has each type of restaurant received from customers

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="green",marker='o')
plt.xlabel("Type of restaurant",c="red", size= 20)
plt.ylabel("votes", c="red",size=20)
```

```python
##conclusion = dining restaurant receices maximum votes
```

```python
#which type of restaurant received more offline orders , so that zomato cam  customers with some good offers?
```

```python
pivot_table = dataframe.pivot_table(index= 'listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu',fmt='d')
plt.title("Heatmap")
plt.xlabel("online order")
plt.ylabel("Listed IN (Type)")
plt.show()
```
